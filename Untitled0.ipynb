```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import io

# 1️⃣ Upload your CSV file
uploaded = files.upload()  # Click "Choose Files" and select your Student_Marks.csv

# 2️⃣ Get the uploaded file name
file_name = list(uploaded.keys())[0]

# 3️⃣ Load dataset
data = pd.read_csv(io.BytesIO(uploaded[file_name]))

# 4️⃣ Show first 5 rows
print("First 5 rows of dataset:")
print(data.head())

# 5️⃣ Show basic statistics
print("\nDataset statistics:")
print(data.describe())

# 6️⃣ Calculate average marks
column_name = 'Marks'  # Your marks column
average_marks = data[column_name].mean()
print("\nAverage marks:", average_marks)

# 7️⃣ Plot histogram of marks
data[column_name].hist()
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```
